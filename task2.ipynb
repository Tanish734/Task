```python
N = int(input('enter the number'))

sum = 0
i = 1

while i < N+1:
    if i % 5 ==0:
        i +=1
        continue
    sum += i

    if sum>300:
        sum -= i
        break
    i +=1

print(sum)
```

```python
sum = 0
count = 0
while True:
    num = int(input('enter the number'))
    if num == 0:
        break
    sum = sum + num
    count = count + 1

print('sum ',sum)
print('avg', sum/count)
```

```python
L = []

for i in range(1000,3001):
    flag = True

    curr = i

    while curr > 0 :
        last = curr%10
        if last % 2 != 0:
            flag = False
            break
        curr = curr//10
    if flag:
        L.append(str(i))

print(",".join(L))
```

```python
pos = [0,0]
while True:
    s = input('Enter the robot path')
    if s == '!':
        break
    direction = s.split()[0]
    steps = int(s.split()[1])

    if direction == 'UP':
        pos[1] = pos[1] + steps
    elif direction == 'DOWN':
        pos[1] = pos[1] - steps
    elif direction == 'LEFT':
        pos[0] = pos[0] - steps
    elif direction == 'RIGHT':
        pos[0] = pos[0] + steps
    else:
        pass

print((pos[0]**2 + pos[1]**2)**0.5)
        
        
```
